```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img_path = 'data\\1.jpg'
```

```python
# • Câu 1: Hãy đọc và hiển thị ảnh có tên 1.jpg trong tập dữ liệu trên
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# • Câu 2: Chuyển ảnh màu thành xám và hiển thị
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# cv2.imshow('img',gray_img)
# Kích thước ảnh
height, width = gray_img.shape

# Kích thước cửa sổ hiển thị mong muốn
desired_width = 800
desired_height = int(height * desired_width / width)

# Resize ảnh để phù hợp với cửa sổ
resized_image = cv2.resize(gray_img, (desired_width, desired_height))

# Hiển thị ảnh
cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Câu 3: Tăng cường độ sáng ảnh màu lên 50 đơn vị, và giảm cường độ sáng xuống
# 80 đơn vị ( Trên ảnh màu gốc ban đầu ) và hiển thị

bright_img = img.astype(np.float32) + 50
bright_img = np.clip(bright_img, 0, 255)
bright_img = bright_img.astype(np.uint8)
plt.imshow(bright_img)
```

```python
bright_img = img.astype(np.float32) - 80
bright_img = np.clip(bright_img, 0, 255)
bright_img = bright_img.astype(np.uint8)
plt.imshow(bright_img)
```
